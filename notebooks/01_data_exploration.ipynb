# CO2 Emission Data Exploration
In this notebook, we will explore the CO2 emissions dataset to get a better understanding of its structure, missing values, and general trends. 

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load the Dataset

```python
# Load the dataset
data_path = 'path_to_your_file/API_EN.ATM.CO2E.KT_DS2_en_csv_v2_32234.csv'
df = pd.read_csv(data_path, skiprows=4)
df.head()
```

## Check for Missing Values

```python
# Check for missing values
df.isnull().sum()
```

## Summary Statistics

```python
# Display summary statistics of the dataset
df.describe()
```

## Visualize Trends Over Time

```python
# Example visualization of CO2 emissions for a country over time
df_country = df[df['Country Name'] == 'United States']
years = df.columns[4:-1]
emissions = df_country[years].T
plt.figure(figsize=(10,6))
plt.plot(years, emissions)
plt.title('CO2 Emissions Over Time (United States)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (kt)')
plt.show()
```
